Import Libraries

```python
import pandas as pd
import numpy as np
```

Load and inspect dataset

```python
df = pd.read_csv(r'C:\Customer Churn Prediction\data\raw\WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

Check missing values

```python
df.isna().sum()
```

Check duplicates

```python
df.duplicated().sum()
```

Get unique values

```python
for i in df.columns:
    print(i,df[i].unique())
    print('*' * 100)
```

Change the datatype of column 'TotalCharges' to numeric

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
```

Check further missing values are present

```python
df.isna().sum()
```

Fill the missing columns with median

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

Strip whitespaces in columns names

```python
df.columns = df.columns.str.strip()
```

Normalize churn labels

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
df.nunique()
```

Rename Columns into  same title format

```python
df.rename(columns = {
    'customerID': 'CustomerID',
    'gender': 'Gender',
    'tenure': 'Tenure',
    }, inplace = True)
```

Create customer table

```python
customers = df[['CustomerID', 'Gender', 'SeniorCitizen', 'Partner', 'Dependents', 'Tenure']]
```

Create service table

```python
services = df[['CustomerID', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','StreamingMovies']]
```

Create billing table

```python
billing = df[['CustomerID', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges']]
```

Create support call table

```python
support_calls = pd.DataFrame()
support_calls['CustomerID'] = df['CustomerID']

support_calls['support_calls'] = (
    (df['TechSupport'] == 'No').astype(int) * np.random.randint(1, 4, size=len(df))
    + (df['InternetService'] == 'Fiber optic').astype(int) * np.random.randint(0, 2, size=len(df))
    + (df['SeniorCitizen']).astype(int) * np.random.randint(0, 2, size=len(df))
)
```

Create churn label

```python
churn_label = df[['CustomerID', 'Churn']]
```

Save all the tables

```python
customers.to_csv("../data/interim/customers.csv", index=False)
services.to_csv("../data/interim/services.csv", index=False)
billing.to_csv("../data/interim/billing.csv", index=False)
support_calls.to_csv("../data/interim/support_calls.csv", index=False)
churn_label.to_csv("../data/interim/churn_label.csv", index=False)
```
